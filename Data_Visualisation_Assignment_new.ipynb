# In this assignment students have to transform iris data into 3 dimensionsand plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
import seaborn as sbn
```

```python
iris_data = sbn.load_dataset('iris')
```

```python
iris_data.head()
```

```python
X = iris_data.iloc[:, :-1].values
y = iris_data.iloc[:, 4].values
```

```python
X[:5, :]
```

```python
y[:5]
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
label_enc_y = LabelEncoder()
y = label_enc_y.fit_transform(y)
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
X
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
X = pca.fit_transform(X)
X
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(1, figsize=(6, 5))
plt.clf()

ax = Axes3D(fig, rect=[0, 0, .9, 1], elev=50, azim=120)
plt.cla()

ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
plt.show()
```
